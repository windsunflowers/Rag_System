```python
import os
import pdfplumber
from typing import List
import re

def load_docs(file_path: str) -> List[str]:
    """读取并切分文档"""
    if not os.path.exists(file_path):
        print(f"错误：找不到文件 {file_path}")
        return []
        
    text = ""
    if file_path.endswith(".pdf"):
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                extracted = page.extract_text()
                if extracted:
                    text += extracted + "\n"
    else:
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read()

    # 先清理并过滤
    chunks = [c.strip() for c in re.split(r'\n+', text) if len(c.strip()) > 5]
    
    # 先打印，再返回
    print(f"成功读取！切分成了 {len(chunks)} 个片段。")
    return chunks
```

```python
import pdfplumber
import re

def split_pdf_into_chunks(pdf_file: str, min_chunk_length=100) -> list[str]:
    full_text = ""

    try:
        with pdfplumber.open(pdf_file) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text:
                    full_text += page_text + "\n"
    except Exception as e:
        print(f"读取PDF失败：{e}")
        return []

    full_text = re.sub(r'\s+', ' ', full_text).strip()

    sentences = re.split(r'(?<=[。！？；])\s+', full_text)

    chunks = []
    current_chunk = ""

    for sentence in sentences:
        sentence = sentence.strip()
        if not sentence:
            continue

        
        if len(current_chunk) + len(sentence) < 800: 
            current_chunk += " " + sentence
        else:
            if current_chunk and len(current_chunk) >= min_chunk_length:
                chunks.append(current_chunk.strip())
            current_chunk = sentence

    if current_chunk and len(current_chunk) >= min_chunk_length:
        chunks.append(current_chunk.strip())

    return chunks

if __name__ == "__main__":
    chunks = split_pdf_into_chunks("doc2.pdf")

    for i, chunk in enumerate(chunks):
        print(f"--- Chunk {i+1} ---")
        print(chunk)
        print("\n")
```

```python
def split_pdf_semantic_chunks(pdf_file, chunk_size=800, overlap=150):
    """
    语义分块 + 重叠
    不会切断句子，不会破坏上下文
    检索精度最高
    """
    full_text = ""

    # 读取PDF
    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            t = page.extract_text()
            if t:
                full_text += t + "\n"

    # 清理多余空白
    full_text = re.sub(r'\s+', ' ', full_text).strip()

    # 先切成完整句子
    sentences = re.split(r'(?<=[。！？;])\s', full_text)

    chunks = []
    current = ""
    for s in sentences:
        s = s.strip()
        if len(current) + len(s) < chunk_size:
            current += " " + s
        else:
            chunks.append(current.strip())
            # 重叠：保留上一块的最后部分
            current = current[-overlap:] + " " + s

    if current:
        chunks.append(current.strip())

    return chunks
if __name__ == "__main__":
    chunks = split_pdf_into_chunks("doc.pdf")

    for i, chunk in enumerate(chunks):
        print(f"--- Chunk {i+1} ---")
        print(chunk)
        print("\n")
```

```python
def split_pdf_accurate_chunks(pdf_file: str, chunk_size=600, overlap=100) -> list[str]:
    """
    PDF精准语义分块：先按章节拆分，再按语义细拆，带章节标题+重叠
    """
    full_text = ""
    # 1. 读取PDF全文
    try:
        with pdfplumber.open(pdf_file) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text:
                    full_text += page_text + "\n"
    except Exception as e:
        print(f"读取PDF失败：{e}")
        return []

    # 2. 清理文本，规范换行
    full_text = re.sub(r'\s+', ' ', full_text).strip()

    # 3. 匹配章节标题，先按大章节拆分
    # 适配你文档的「第X章」格式，可自行修改正则
    chapter_pattern = re.compile(r'(第[一二三四五六七八九十百]+章：[^\s]+)')
    chapter_splits = chapter_pattern.split(full_text)

    # 结构化章节：{章节标题: 章节正文}
    chapters = {}
    current_chapter = "前言"
    for part in chapter_splits:
        if chapter_pattern.match(part):
            current_chapter = part
        else:
            chapters[current_chapter] = chapters.get(current_chapter, "") + part

    # 4. 章节内细拆成语义块，带章节标题前缀
    final_chunks = []
    for chapter_title, chapter_content in chapters.items():
        # 先按句子拆分，保证语义完整
        sentences = re.split(r'(?<=[。！？；])\s+', chapter_content)
        current_chunk = ""

        for sentence in sentences:
            sentence = sentence.strip()
            if not sentence:
                continue
            # 累加句子，达到目标长度就生成块
            if len(current_chunk) + len(sentence) < chunk_size:
                current_chunk += " " + sentence
            else:
                # 给块加上章节标题前缀
                if current_chunk:
                    final_chunks.append(f"【{chapter_title}】{current_chunk.strip()}")
                # 保留重叠内容，避免语义断裂
                current_chunk = current_chunk[-overlap:] + " " + sentence

        # 处理章节最后一个块
        if current_chunk:
            final_chunks.append(f"【{chapter_title}】{current_chunk.strip()}")

    return final_chunks


# 调用示例
if __name__ == "__main__":
    chunks = split_pdf_accurate_chunks("doc.pdf")
    # 打印结果
    for i, chunk in enumerate(chunks):
        print(f"--- Chunk {i+1} ---\n{chunk}\n")
```
