```python
import os
import pdfplumber
from typing import List
import re

def load_docs(file_path: str) -> List[str]:
    """读取并切分文档"""
    if not os.path.exists(file_path):
        print(f"错误：找不到文件 {file_path}")
        return []
        
    text = ""
    if file_path.endswith(".pdf"):
        with pdfplumber.open(file_path) as pdf:
            for page in pdf.pages:
                extracted = page.extract_text()
                if extracted:
                    text += extracted + "\n"
    else:
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read()

    # 先清理并过滤
    chunks = [c.strip() for c in re.split(r'\n+', text) if len(c.strip()) > 5]
    
    # 先打印，再返回
    print(f"成功读取！切分成了 {len(chunks)} 个片段。")
    return chunks
```

```python
my_chunks = load_docs("doc.pdf") 
for i, chunks in enumerate(my_chunks):
    print(f"my_chunks {i}")
    print(my_chunks)
```

```python
import pdfplumber
import re

def split_pdf_into_chunks(pdf_file: str, min_chunk_length=100) -> list[str]:
    full_text = ""

    try:
        with pdfplumber.open(pdf_file) as pdf:
            for page in pdf.pages:
                page_text = page.extract_text()
                if page_text:
                    full_text += page_text + "\n"
    except Exception as e:
        print(f"读取PDF失败：{e}")
        return []

    full_text = re.sub(r'\s+', ' ', full_text).strip()

    sentences = re.split(r'(?<=[。！？；])\s+', full_text)

    chunks = []
    current_chunk = ""

    for sentence in sentences:
        sentence = sentence.strip()
        if not sentence:
            continue

        
        if len(current_chunk) + len(sentence) < 800: 
            current_chunk += " " + sentence
        else:
            if current_chunk and len(current_chunk) >= min_chunk_length:
                chunks.append(current_chunk.strip())
            current_chunk = sentence

    if current_chunk and len(current_chunk) >= min_chunk_length:
        chunks.append(current_chunk.strip())

    return chunks

if __name__ == "__main__":
    chunks = split_pdf_into_chunks("doc2.pdf")

    for i, chunk in enumerate(chunks):
        print(f"--- Chunk {i+1} ---")
        print(chunk)
        print("\n")
```
